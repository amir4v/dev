```python
def pcc(x, y):
    """Pearson Correlation Coefficient"""
    SP = ((x - x.mean()) * (y - y.mean())).sum()
    SSx = ((x - x.mean()) ** 2).sum()
    SSy = ((y - y.mean()) ** 2).sum()
    
    return SP / math.sqrt(SSx * SSy)
```

```python
pcc(
    np.array([8, 11, -5]),
    np.array([4, 5, 6]),
)
```
